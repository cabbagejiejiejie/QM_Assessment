```python
import os
import numpy as np
import pandas as pd
```

```python
path_household_deprivation = 'https://github.com/cabbagejiejiejie/QM_Assessment/blob/main/data/Household%20deprivation.xlsx?raw=true'
sheet_name = '2021'

df_household_deprivation = pd.read_excel(path_household_deprivation, sheet_name=sheet_name, engine='openpyxl')
```

```python
df_household_deprivation
```

```python
grouped = df_household_deprivation.groupby('local authority name')

list_split_dfs = []

for group_name, group_df in grouped:
    list_split_dfs.append(group_df.reset_index(drop=True))
```

```python
for dfs in list_split_dfs:
    sum_row = dfs.sum()
    dfs.loc['total'] = sum_row
    dfs['poor_household'] = dfs['1 dimension']/dfs['All Households']
    dfs.at['total', 'ward code'] = ''
    dfs.at['total', 'ward name'] = ''
    dfs.at['total', 'local authority code'] = dfs.at[0, 'local authority code']
    dfs.at['total', 'local authority name'] = dfs.at[0, 'local authority name']
```
